Generating negatives

```python
import pandas as pd
import random

#Load positive epitope data

epitope_df = training_data
positive_seqs = epitope_df["Epitope"].dropna().unique().tolist()
positive_lens = [len(seq) for seq in positive_seqs]
positive_set = set(positive_seqs)

# Load proteome Fasta
def parse_fasta(filepath):
    sequences = []
    with open(filepath, 'r') as f:
        seq_id = None
        seq = []
        for line in f:
            line = line.strip()
            if line.startswith(">"):
                if seq:
                    sequences.append("".join(seq))
                    seq = []
            else:
                seq.append(line)
        if seq:
            sequences.append("".join(seq))
    return sequences

proteins = parse_fasta("uniprotkb_proteome_UP000001584_2025_04_12.fasta")

# Negative samples
def generate_negatives(proteins, lengths, positives, max_attempts=10000):
    negatives = []
    attempts = 0
    all_prots = "".join(proteins)
    while len(negatives) < len(lengths) and attempts < max_attempts:
        length = lengths[len(negatives)]
        prot = random.choice(proteins)
        if len(prot) < length:
            continue
        start = random.randint(0, len(prot) - length)
        candidate = prot[start:start + length]
        if candidate not in positives:
            negatives.append(candidate)
        attempts += 1
    return negatives

negative_seqs = generate_negatives(proteins, positive_lens, positive_set)

# Combining in a dataset
data = pd.DataFrame({
    "sequence": positive_seqs + negative_seqs,
    "label": [1] * len(positive_seqs) + [0] * len(negative_seqs)
})

# Shuffle and save
data = data.sample(frac=1).reset_index(drop=True)
data.to_csv("epitope_dataset.csv", index=False)
print("✅ Dataset saved as 'epitope_dataset.csv' with {} samples.".format(len(data)))
```

Amino acid Content

```python
import pandas as pd
from collections import Counter


# Define the 20 standard amino acids
AMINO_ACIDS = "ACDEFGHIKLMNPQRSTVWY"

# Function to calculate AAC
def compute_aac(seq):
    counts = Counter(seq)
    total = len(seq)
    return [counts[aa] / total for aa in AMINO_ACIDS]

# Apply to all sequences
aac_features = desc_df["peptide"].apply(compute_aac)
aac_df = pd.DataFrame(aac_features.tolist(), columns=[f"AAC_{aa}" for aa in AMINO_ACIDS])

```

```python
bcell_data = pd.concat([desc_df, aac_df], axis=1)
```

Model testing on present Features


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler


# Split into features and labels
X = bcell_data.drop(columns=["peptide", "label"])
y = bcell_data["label"]

# Split before SMOTE (to avoid data leakage)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.2, random_state=42
)

# Standardize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Apply SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train_scaled, y_train)


# Train Random Forest
model = RandomForestClassifier(n_estimators=500, random_state=42)
model.fit(X_resampled, y_resampled)

# Predict on test set
y_pred = model.predict(X_test_scaled)

# Evaluation
print(classification_report(y_test, y_pred))
```

```python
#F1-score
from sklearn.model_selection import StratifiedKFold, cross_val_score
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(model, X, y, cv=cv, scoring='f1')
print("Average F1:", scores.mean())
```

Adding Dipeptide Features


```python
from collections import Counter
import pandas as pd
import itertools

amino_acids = 'ACDEFGHIKLMNPQRSTVWY'
dipeptides = [''.join(p) for p in itertools.product(amino_acids, repeat=2)]

def compute_dipeptide_composition(sequence):
    seq = sequence.upper()
    counts = Counter([seq[i:i+2] for i in range(len(seq)-1) if seq[i] in amino_acids and seq[i+1] in amino_acids])
    total = sum(counts.values())
    return [counts[dp] / total if total > 0 else 0 for dp in dipeptides]

# Apply on your DataFrame
df_dipep = bcell_data["peptide"].apply(compute_dipeptide_composition)
df_dipep = pd.DataFrame(df_dipep.tolist(), columns=dipeptides)


```
